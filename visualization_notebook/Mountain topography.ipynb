##### Construct a contour map to display the topography of a mountain region, with 'sandybrown' indicating lower altitudes and 'skyblue' indicating higher altitudes.

## **Data:** 

## **Solution:** 

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(42)

x = np.linspace(-5, 5, 100)  
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
R = np.sqrt(X**2 + Y**2)
Z = 1000 - 200 * R 

Z += np.random.normal(0, 50, Z.shape)

plt.figure(figsize=(12, 8))
contour = plt.contourf(X, Y, Z, cmap='coolwarm')
cbar = plt.colorbar(contour)
cbar.set_label('Altitude (meters)')

plt.contourf(X, Y, Z, levels=[-1000, -500, 0, 500, 1000], colors=['sandybrown', 'peachpuff', 'lightgreen', 'skyblue'])

plt.title('Topography of a Mountain Region')
plt.xlabel('X Coordinate')
plt.ylabel('Y Coordinate')
plt.show()
```
