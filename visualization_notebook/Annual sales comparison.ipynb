##### Construct a radial bar chart to compare annual sales performance of different automotive brands, using 'gold' for Brand A, 'silver' for Brand B, and 'bronze' for Brand C.

## **Data:** 

```python
import pandas as pd 

data = [
    {
        "brand": "Brand A",
        "sales": 350000
    },
    {
        "brand": "Brand B",
        "sales": 450000
    },
    {
        "brand": "Brand C",
        "sales": 300000
    }
]

df = pd.DataFrame(data)
```

## **Solution:** 

```python
import matplotlib.pyplot as plt
import numpy as np

# Data preparation
labels = df['brand']
sales = df['sales']
num_vars = len(labels)

# Compute angle for each bar
angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()

# Repeat the first value to close the circle for the sales data
sales = np.concatenate((sales, [sales[0]]))
angles += angles[:1]

# Create the figure
fig, ax = plt.subplots(figsize=(8, 8), subplot_kw=dict(polar=True))
colors = ['gold', 'silver', '#cd7f32']  # '#cd7f32' is the hex code for bronze

# Plot bars
bars = ax.bar(angles, sales, color=colors, edgecolor='black', linewidth=2, alpha=0.75)

# Add labels
ax.set_yticklabels([])
ax.set_xticks(angles[:-1])
ax.set_xticklabels(labels)

# Add a title
plt.title('Annual Sales Performance of Automotive Brands', size=20, color='black', y=1.1)

plt.show()
```
