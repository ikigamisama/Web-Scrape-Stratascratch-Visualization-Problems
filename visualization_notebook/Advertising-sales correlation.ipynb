##### Plot a bubble chart to explore the correlation between advertising spend and sales volume across different media channels, using 'chocolate' for TV, 'cyan' for online, and 'crimson' for print media.

## **Data:** 

```python
import pandas as pd 

data = [
    {
        "media_channel": "TV",
        "ad_spend": 71615,
        "sales_volume": 684
    },
    {
        "media_channel": "Online",
        "ad_spend": 68564,
        "sales_volume": 445
    },
    {
        "media_channel": "Print",
        "ad_spend": 59689,
        "sales_volume": 466
    },
    {
        "media_channel": "TV",
        "ad_spend": 20742,
        "sales_volume": 654
    },
    {
        "media_channel": "Online",
        "ad_spend": 63459,
        "sales_volume": 830
    },
    {
        "media_channel": "Print",
        "ad_spend": 76455,
        "sales_volume": 267
    },
    {
        "media_channel": "TV",
        "ad_spend": 89463,
        "sales_volume": 168
    },
    {
        "media_channel": "Online",
        "ad_spend": 66718,
        "sales_volume": 532
    },
    {
        "media_channel": "Print",
        "ad_spend": 78119,
        "sales_volume": 995
    }
]

df = pd.DataFrame(data)
```

## **Solution:** 

```python
import matplotlib.pyplot as plt

# Colors for each media channel
colors = {'TV': 'chocolate', 'Online': 'cyan', 'Print': 'crimson'}

# Plotting
plt.figure(figsize=(10, 6))
for channel, group in df.groupby('media_channel'):
    plt.scatter(group['ad_spend'], group['sales_volume'], s=group['sales_volume']*10, color=colors[channel], alpha=0.5, label=channel)

plt.title('Correlation Between Advertising Spend and Sales Volume by Media Channel')
plt.xlabel('Advertising Spend ($)')
plt.ylabel('Sales Volume (units)')
plt.legend(title='Media Channel')
plt.grid(True)
plt.show()
```
