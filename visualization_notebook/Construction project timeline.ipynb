##### Develop a Gantt chart to visualize the timeline for different phases of a construction project, using 'olive' for planning, 'darkorange' for operation, and 'cornflowerblue' for assessment.

## **Data:** 

```python
import pandas as pd 

data = [
    {
        "task": "Planning",
        "start": "2023-01-01",
        "finish": "2023-01-31",
        "phase": "Planning"
    },
    {
        "task": "operation",
        "start": "2023-02-01",
        "finish": "2023-05-31",
        "phase": "operation"
    },
    {
        "task": "Assessment",
        "start": "2023-06-01",
        "finish": "2023-07-31",
        "phase": "Assessment"
    }
]

df = pd.DataFrame(data)
```

## **Solution:** 

```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
#plotting the chart
colors = {'Planning': 'olive', 'operation': 'darkorange', 'Assessment': 'cornflowerblue'}
fig, ax = plt.subplots(figsize=(12, 8))

df['start'] = pd.to_datetime(df['start'])
df['finish'] = pd.to_datetime(df['finish'])

for i, task in df.iterrows():
    ax.barh(task['task'], (task['finish'] - task['start']).days, left=task['start'], color=colors[task['phase']])

ax.set_xlabel('Date')
ax.set_ylabel('Task')
ax.set_title('Gantt Chart for Construction Project Phases')

ax.xaxis_date()
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
plt.xticks(rotation=45)

legend_elements = [plt.Line2D([0], [0], color='olive', lw=4, label='Planning'),
                   plt.Line2D([0], [0], color='darkorange', lw=4, label='operation'),
                   plt.Line2D([0], [0], color='cornflowerblue', lw=4, label='Assessment')]
ax.legend(handles=legend_elements, loc='upper right')

plt.show()
```
