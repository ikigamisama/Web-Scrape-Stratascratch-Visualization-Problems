##### Create a bar chart to compare the annual revenue of different retail brands, using 'gold' for Brand A, 'silver' for Brand B, and 'bronze' for Brand C.

## **Data:** 

```python
import pandas as pd 

data = [
    {
        "brand": "Brand A",
        "annual_revenue": 500
    },
    {
        "brand": "Brand B",
        "annual_revenue": 300
    },
    {
        "brand": "Brand C",
        "annual_revenue": 200
    }
]

df = pd.DataFrame(data)
```

## **Solution:** 

```python
import matplotlib.pyplot as plt

colors = {'Brand A': 'gold', 'Brand B': 'silver', 'Brand C': '#cd7f32'} 
plt.figure(figsize=(10, 6))
bars = plt.bar(df['brand'], df['annual_revenue'], color=[colors[brand] for brand in df['brand']])
plt.title('Annual Revenue of Different Retail Brands')
plt.xlabel('Brand')
plt.ylabel('Annual Revenue ($ Millions)')
plt.grid(axis='y')
plt.show()
```
