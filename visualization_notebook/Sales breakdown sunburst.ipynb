##### Utilize a sunburst chart to break down the sales data of a company by region and product type, with 'navajowhite' for North America, 'sandybrown' for Europe, and 'lightsteelblue' for Asia.

## **Data:** 

```python
import pandas as pd 

data = [
    {
        "region": "North America",
        "product_type": "Electronics",
        "sales": 100000
    },
    {
        "region": "North America",
        "product_type": "Apparel",
        "sales": 50000
    },
    {
        "region": "North America",
        "product_type": "Toys",
        "sales": 25000
    },
    {
        "region": "Europe",
        "product_type": "Electronics",
        "sales": 80000
    },
    {
        "region": "Europe",
        "product_type": "Apparel",
        "sales": 40000
    },
    {
        "region": "Europe",
        "product_type": "Toys",
        "sales": 20000
    },
    {
        "region": "Asia",
        "product_type": "Electronics",
        "sales": 90000
    },
    {
        "region": "Asia",
        "product_type": "Apparel",
        "sales": 30000
    },
    {
        "region": "Asia",
        "product_type": "Toys",
        "sales": 10000
    }
]

df = pd.DataFrame(data)
```

## **Solution:** 

```python
import matplotlib.pyplot as plt

region_colors = {'North America': 'navajowhite', 'Europe': 'sandybrown', 'Asia': 'lightsteelblue'}
df['colors'] = df['region'].map(region_colors)

fig, ax = plt.subplots()
size = 0.3

vals = df.groupby('region')['sales'].sum()
cmap = plt.get_cmap("tab20c")
outer_colors = df['region'].map(region_colors).unique()
ax.pie(vals, radius=1, colors=outer_colors, wedgeprops=dict(width=size, edgecolor='w'),
       labels=vals.index, autopct='%1.1f%%', pctdistance=0.85)


vals = df['sales']
inner_colors = df['colors']
ax.pie(vals, radius=1-size, colors=inner_colors, wedgeprops=dict(width=size, edgecolor='w'),
       labels=df['product_type'], labeldistance=0.75, autopct='%1.1f%%', pctdistance=0.75)

centre_circle = plt.Circle((0,0),1-size,fc='white')
fig.gca().add_artist(centre_circle)

ax.set_aspect('equal')
plt.title('Sales Breakdown by Region and Product Type')
plt.show()
```
